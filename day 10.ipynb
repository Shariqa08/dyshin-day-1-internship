```python
import pandas as pd

# Create a sample dataset
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, None, 30, 22],
    'Salary': [50000, 60000, None, 45000]
}

df = pd.DataFrame(data)

# ✅ Fixed: Handle missing values (no inplace on chained assignment)
df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Salary'] = df['Salary'].fillna(df['Salary'].median())

# Rename columns
df.rename(columns={'Name': 'Employee', 'Salary': 'Annual_Salary'}, inplace=True)

# Filter data
high_salary = df[df['Annual_Salary'] > 48000]

# Sort data
df_sorted = df.sort_values(by='Annual_Salary', ascending=False)

# Add new column
df['Monthly_Salary'] = df['Annual_Salary'] / 12

print(df)
print("\nHigh Salary Employees:\n", high_salary)
print("\nSorted Data:\n", df_sorted)
```

```python
# Variables and Data Types
name = "Alice"
age = 25
salary = 50000.0
is_employed = True

print(type(name), type(age), type(salary), type(is_employed))

# Conditional Statements
if age >= 18:
    print(f"{name} is an adult.")
else:
    print(f"{name} is a minor.")

# Loops
numbers = [1, 2, 3, 4, 5]
total = 0
for num in numbers:
    total += num
print("Sum:", total)

# Functions
def calculate_bonus(salary, percent):
    return salary * (percent / 100)

bonus = calculate_bonus(salary, 10)
print(f"Bonus: {bonus}")

# List Comprehension
squares = [x**2 for x in range(1, 6)]
print("Squares:", squares)

# Dictionary
employee = {
    'name': 'Alice',
    'age': 25,
    'salary': 50000
}
for key, value in employee.items():
    print(f"{key}: {value}")
```
