```python
# Boolean values example

is_student = True
is_graduated = False

print("Is the person a student?", is_student)
print("Is the person graduated?", is_graduated)
```

```python
# Using Boolean values in a condition

age = 20
is_adult = age >= 18

print("Is the person adult?", is_adult)
```

```python
# Creating a list

fruits = ["apple", "banana", "mango", "orange"]

print("Fruits list:", fruits)
print("First fruit:", fruits[0])
print("Total fruits:", len(fruits))
```

```python
# Adding items to list

numbers = [10, 20, 30]

numbers.append(40)

print("Updated list:", numbers)
```

```python
# Creating a tuple

colors = ("red", "green", "blue")

print("Tuple values:", colors)
print("First color:", colors[0])
```

```python
# Creating a set

numbers = {1, 2, 3, 4, 4, 5}

print("Set values:", numbers)
```

```python
# Comparing different data structures

my_list = [1, 2, 3]
my_tuple = (1, 2, 3)
my_set = {1, 2, 3}

print("List:", my_list)
print("Tuple:", my_tuple)
print("Set:", my_set)
```
