```python
import pandas as pd
```

```python
data = {
    'employee': ['A', 'B', 'C', 'D', 'E'],
    'commute_time': ['30 mins', '1 hour', '45 minutes', '2 hours', 'approx 20 min']
}

df = pd.DataFrame(data)

df.head()
```

```python
def convert_to_minutes(time):
    time = str(time).lower()
    
    if 'hour' in time:
        num = float(time.split()[0])
        return num * 60
    
    elif 'min' in time:
        num = float(time.split()[0])
        return num
    
    else:
        return None
```

```python
import pandas as pd
import re

def convert_to_minutes(time):
    if pd.isna(time):
        return None
    
    time = str(time).lower()
    
    # extract number from text
    num = re.findall(r'\d+', time)
    
    if not num:
        return None
    
    num = float(num[0])
    
    if 'hour' in time:
        return num * 60
    
    elif 'min' in time:
        return num
    
    else:
        return None
```

```python
df['commute_mins'] = df['commute_time'].apply(convert_to_minutes)
```
