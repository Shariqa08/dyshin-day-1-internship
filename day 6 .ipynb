```python
# defining a simple function
def greet():
    print("Hello! Welcome to Python functions")

# calling the function
greet()
```

```python
# function with parameters
def greet_user(name):
    print("Hello", name)

greet_user("Shariqa")
```

```python
# function that returns a value
def add_numbers(a, b):
    result = a + b
    return result

sum_value = add_numbers(10, 5)
print("Sum:", sum_value)
```

```python
def check_even_odd(num):
    if num % 2 == 0:
        return "Even Number"
    else:
        return "Odd Number"

print(check_even_odd(7))
```

```python
def square(number):
    return number * number

print("Square:", square(6))
```

```python
def student_info(name, age, course):
    print("Name:", name)
    print("Age:", age)
    print("Course:", course)

student_info("Shariqa", 21, "Data Science")
```

```python
def find_max(a, b, c):
    return max(a, b, c)

print("Maximum value:", find_max(10, 25, 15))
```
