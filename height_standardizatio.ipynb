```python
import pandas as pd
import re
```

```python
data = {
    "Name": ["A", "B", "C", "D", "E"],
    "Height": ["170 cm", "5'7\"", "5.5 ft", "160", "6 ft"]
}

df = pd.DataFrame(data)

df
```

```python
def convert_height_to_cm(height):
    
    if pd.isna(height):
        return None
    
    height = str(height).lower().strip()

    # Case 1: centimeters
    if "cm" in height:
        num = re.findall(r'\d+', height)
        return float(num[0])

    # Case 2: feet and inches (5'7")
    elif "'" in height:
        parts = re.findall(r'\d+', height)
        feet = float(parts[0])
        inches = float(parts[1]) if len(parts) > 1 else 0
        return round((feet * 30.48) + (inches * 2.54), 2)

    # Case 3: decimal feet (5.5 ft)
    elif "ft" in height:
        num = re.findall(r'\d+\.?\d*', height)
        return round(float(num[0]) * 30.48, 2)

    # Case 4: numeric value assumed cm
    elif height.isdigit():
        return float(height)

    else:
        return None
```

```python
df["Height_cm"] = df["Height"].apply(convert_height_to_cm)
```
